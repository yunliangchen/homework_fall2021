# Part 1: "Unsupervised" RND and exploration performance.

```python
def get_summary_df(ea):
    result = pd.DataFrame(ea.Scalars('Eval_AverageReturn'))
    result = result.rename(columns={"value": "Eval_AverageReturn"})
#     result['Train_BestReturn'] = [result.Train_AverageReturn[0]] + [e.value for e in ea.Scalars('Train_BestReturn')]
#     result['Train_AverageReturn'] = [e.value for e in ea.Scalars('Train_AverageReturn')]    
#     result['Train_BestReturn'] = [e.value for e in ea.Scalars('Train_BestReturn')]
    return result
```

```python
# PointmassEasy-v0
ea_q1_env1_random_PointmassEasy = event_accumulator.EventAccumulator('data/hw5_expl_q1_env1_random_PointmassEasy-v0_13-11-2021_22-48-50/events.out.tfevents.1636872530.triton2')
# ea_q1_env1_rnd_PointmassEasy = event_accumulator.EventAccumulator('data/hw5_expl_q1_env1_rnd_PointmassEasy-v0_13-11-2021_22-28-41/events.out.tfevents.1636871321.triton2')
ea_q1_env1_rnd_PointmassEasy = event_accumulator.EventAccumulator('data/hw5_expl_q1_env1_rnd_PointmassEasy-v0_14-11-2021_01-38-41/events.out.tfevents.1636882721.triton2')
ea_list = [ea_q1_env1_random_PointmassEasy, ea_q1_env1_rnd_PointmassEasy]
ea_list = [ea.Reload() for ea in ea_list]
ea_q1_env1_random_PointmassEasy.Tags()
```

```python
result_list = list(map(get_summary_df, ea_list))
for result in result_list:
    display(result)
```

```python
labels_list = ['q1_env1_random_PointmassEasy', 'q1_env1_rnd_PointmassEasy']
fig = plt.figure()
for i in range(2):
    plt.plot(result_list[i].step, result_list[i].Eval_AverageReturn, linestyle='-', label = labels_list[i])
plt.legend()
plt.title('RND vs Random Exploration on PointmassEasy-v0')
plt.xlabel('Number of Iterations')
plt.ylabel('Average per-epoch Reward')
# plt.ticklabel_format(axis="x", style="sci", scilimits=(0,0))
plt.show()
```
